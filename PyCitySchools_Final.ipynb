```python
# Dependencies and Setup
import os
import pandas as pd

    
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
district_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Define Columns
school_count = school_data["school_name"].count()  # total number of schools
student_count = district_data["student_name"].count() # total number of students
total_budget = school_data["budget"].sum() # total budget
dist_avg_math = district_data["math_score"].mean() # average math score
dist_avg_reading = district_data["reading_score"].mean() # average reading score
#------------------------------------------------------------------------------------------------
# number of rows in which value of 'math_score' column is 70 or greater:
seriesObj1 = district_data.apply(lambda x: True if x['math_score'] >= 70 else False , axis=1)
pass_math = len(seriesObj1[seriesObj1 == True].index)  # Count number of True in series
pct_pass_math = pass_math / student_count * 100
#------------------------------------------------------------------------------------------------ 
# number of rows in which value of 'reading_score' column is 70 or greater:
seriesObj2 = district_data.apply(lambda x: True if x['reading_score'] >= 70 else False , axis=1)
pass_reading = len(seriesObj2[seriesObj2 == True].index)  # Count number of True in series
pct_pass_reading = pass_reading / student_count * 100
#------------------------------------------------------------------------------------------------
overall_avg_score = (dist_avg_math + dist_avg_reading) / 2  # overall average score
```

```python
# Create dataframe (District Summary Table - DST)
DST = pd.DataFrame({"Total Number of Schools": [school_count],
                                   "Total Number of Students": [student_count],
                                   "Total Budget": [total_budget],              
                                   "Average Math Score": [dist_avg_math],
                                   "Average Reading Score": [dist_avg_reading],
                                   "Percent Passing Math": [pct_pass_math],
                                   "Percent Passing Reading": [pct_pass_reading],
                                   "Percent Overall Passing Rate": [overall_avg_score]})
DST
DST.style.format({"Total Budget": "${:,.2f}"})
```

  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Define columns   (School Summary Table = SST)
school_name = school_data["school_name"]
total_students = school_data["size"]                              #students sorted by index
a = district_data.groupby(['school_name']).groups.keys()

grouped_school_data = district_data.set_index("school_name").groupby(['school_name'])  # gets school names
school_type = school_data.set_index('school_name')['type']                  # gets type of school
students_per_school = district_data.groupby('school_name')['size'].count()  # gets student count by high school
school_budget = district_data.groupby('school_name')['budget'].unique()    # gets budget
per_student_budget = school_budget / students_per_school                  # Per Student Budget
school_avg_math = district_data.groupby('school_name')['math_score'].mean()           #gets avg math score
school_avg_reading = district_data.groupby('school_name')['reading_score'].mean()     #gets avg reading score

students_passing_math = district_data[district_data["math_score"] >= 70].groupby("school_name")["student_name"].count()
pct_pass_math_school = students_passing_math / students_per_school * 100

students_passing_reading = district_data[district_data["reading_score"] >= 70].groupby("school_name")["student_name"].count()
pct_pass_reading_school = students_passing_reading / students_per_school * 100

overall_pass_rate = (pct_pass_reading_school + pct_pass_math_school) / 2
```

```python
SST = pd.DataFrame({"School Type": school_type,                      # Huang High School        District
                    "Total Students": students_per_school,           # Bailey High School       4976
                    "Total School Budget": school_budget,            # Bailey High School       [3124928]
                    "Per Student Budget": per_student_budget,        # Bailey High School       [628.0]
                    "Average Math Score": school_avg_math,           # Bailey High School       77.048432
                    "Average Reading Score": school_avg_reading,     # Bailey High School       81.033963
                    "% Passing Math": pct_pass_math_school,          # Bailey High School       66.680064
                    "% Passing Reading": pct_pass_reading_school,    # Bailey High School       81.933280
                    "% Overall Passing Rate": overall_pass_rate})    # Bailey High School       74.306672


#SST = SST[["School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score", 
#                   "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
# ADD FORMATTING HERE
SST
```
